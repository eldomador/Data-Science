```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()  # an empty figure with no axes
fig.suptitle('No axes on this figure')  # Add a title so we know which it is

fig, ax_lst = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
```

```python
x = np.linspace(0, 4, 100)

plt.plot(x, x, label='funkcja liniowa')
plt.plot(x, x**2, label='funkcja kwadratowa')
plt.plot(x, x**3, label='funkcja sześcienna')

plt.xlabel('X')
plt.ylabel('Y')

plt.title("Wykres")

plt.legend()

plt.show()
```

```python
x = np.arange(0, 10, 0.2)
y = np.cos(x)
fig, ax = plt.subplots()
ax.plot(x, y)
plt.title("Cos")
plt.show()
```

```python
import numpy as np
import pandas as pd
```

```python
fury=pd.read_csv('samochody1tys.csv') 
```

```python
fury
```

```python
cenaMax = fury['cena'].max()
```

```python
cenaMin = fury['cena'].min()
```

```python
rokProdukcjiMin = fury['rok_produkcji'].min()
```

```python
rokProdukcjiMax = fury['rok_produkcji'].max()
```

```python
przebiegMin = fury['przebieg'].min()
```

```python
przebiegMax = fury['przebieg'].max()
```

```python
objects = ('Cena Max', 'Cena Min')
y_pos = np.arange(len(objects))
performance = [cenaMax,cenaMin]

plt.bar(y_pos, performance, align='center', alpha=1)
plt.xticks(y_pos, objects)
plt.ylabel('Ilość')
plt.title('Cena')

plt.show()
```

```python
objects = ('Rok Produkcji Min', 'Rok Produkcji Max')
y_pos = np.arange(len(objects))
performance = [rokProdukcjiMin,rokProdukcjiMax]

plt.bar(y_pos, performance, align='center', alpha=1)
plt.xticks(y_pos, objects)
plt.ylabel('Ilość')
plt.title('Rok Produkcji')

plt.show()
```

```python
objects = ('Przebieg Min', 'Przebieg Max')
y_pos = np.arange(len(objects))
performance = [przebiegMin,przebiegMax]

plt.bar(y_pos, performance, align='center', alpha=1)
plt.xticks(y_pos, objects)
plt.ylabel('Ilość')
plt.title('Przebieg')

plt.show()
```

```python
Audi = fury.loc[fury['marka'] == 'Audi']['marka'].count()
```

```python
Bmw = fury.loc[fury['marka'] == 'BMW']['marka'].count()
```

```python
Fiat = fury.loc[fury['marka'] == 'Fiat']['marka'].count()
```

```python
Ford = fury.loc[fury['marka'] == 'Ford']['marka'].count()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Audi', 'Bmw', 'Fiat', 'Ford'
sizes = [Audi, Bmw, Fiat, Ford]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
